```python
import pandas as pd
import matplotlib.pyplot as plt

print("======== STUDENT RESULT ANALYSIS SYSTEM ========")

# CSV file load
data = pd.read_csv("/content/students_50.csv")

# Total calculate
data["Total"] = data[["Maths","English","Computer","SST","Reasoning"]].sum(axis=1)

# Percentage calculate
data["Percentage"] = data["Total"] / 5

# Result pass/fail
data["Result"] = data["Percentage"].apply(lambda x: "Pass" if x >= 40 else "Fail")

# Roll number input
roll = int(input("\nEnter Student Roll Number: "))

# Student search
student = data[data["Roll"] == roll]

if student.empty:
    print("\nStudent Not Found in Database")

else:

    s = student.iloc[0]

    print("\n------ STUDENT RESULT ------\n")

    print("Name       :", s["Name"])
    print("Roll       :", s["Roll"])
    print("Maths      :", s["Maths"])
    print("English    :", s["English"])
    print("Computer   :", s["Computer"])
    print("SST        :", s["SST"])
    print("Reasoning  :", s["Reasoning"])
    print("Total      :", s["Total"])
    print("Percentage :", s["Percentage"])
    print("Result     :", s["Result"])

    # Subject graph
    subjects = ["Maths","English","Computer","SST","Reasoning"]
    marks = [s["Maths"],s["English"],s["Computer"],s["SST"],s["Reasoning"]]

    plt.figure()
    plt.bar(subjects, marks)
    plt.title("Student Subject Performance")
    plt.xlabel("Subjects")
    plt.ylabel("Marks")
    plt.show()

# Category analysis
topper = data[data["Percentage"] >= 80]
average = data[(data["Percentage"] >= 60) & (data["Percentage"] < 80)]
weak = data[(data["Percentage"] >= 40) & (data["Percentage"] < 60)]
fail = data[data["Percentage"] < 40]

plt.figure()
plt.bar(topper["Name"], topper["Percentage"])
plt.title("Topper Students (80%+)")
plt.xticks(rotation=90)
plt.show()

plt.figure()
plt.bar(average["Name"], average["Percentage"])
plt.title("Average Students (60-80%)")
plt.xticks(rotation=90)
plt.show()

plt.figure()
plt.bar(weak["Name"], weak["Percentage"])
plt.title("Weak Students (40-60%)")
plt.xticks(rotation=90)
plt.show()

plt.figure()
plt.bar(fail["Name"], fail["Percentage"])
plt.title("Fail Students (<40%)")
plt.xticks(rotation=90)
plt.show()
```
